```python
import utils
import numpy as np
from sklearn.metrics.pairwise import pairwise_distances, cosine_similarity
from sklearn.metrics import accuracy_score, confusion_matrix, mean_squared_error
import random
```

```python
X_train_full, y_train_full = utils.load_train_data("Visit")
```

```python
X_test = utils.load_test_data("Visit")
```

```python
users = list(set(X_train_full['user'].tolist()))
items = list(set(X_train_full['item'].tolist()))
u_map = dict(zip(users, range(0,len(users))))
i_map = dict(zip(items, range(0,len(items))))
```

```python
train_data_matrix = np.zeros((len(users), len(items)))
```

```python
train = list(zip(X_train_full.values.tolist(), y_train_full))
```

```python
for x, y in train:
    train_data_matrix[u_map[x[0]], i_map[x[1]]] = y
```

```python
item_similarity = cosine_similarity(train_data_matrix.T, dense_output=False)
```

```python
def predict(ratings, similarity, type='user'):
    if type == 'user':
        mean_user_rating = ratings.mean(axis=1)
        ratings_diff = (ratings - mean_user_rating[:, np.newaxis])
        pred = mean_user_rating[:, np.newaxis] + similarity.dot(ratings_diff) / np.array([np.abs(similarity).sum(axis=1)]).T
    elif type == 'item':
        pred = train_data_matrix.dot(similarity) / np.array([np.abs(similarity).sum(axis=1)])
    return pred
```

```python
item_similarity = 1-item_similarity
```

```python
item_prediction = predict(train_data_matrix, item_similarity, type='item')
```
